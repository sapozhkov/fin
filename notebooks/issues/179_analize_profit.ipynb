```python
import pandas as pd

# Чтение CSV-файла в DataFrame
file_path = 'runs.csv'  # Укажите путь к вашему файлу
df = pd.read_csv(file_path)

# Вывод первых нескольких строк для проверки
print(df.head())
```

```python
columns = df.columns.tolist()
print(columns)
```

```python
columns_to_keep = ['id', 'instrument', 'date', 'profit', 'instrument_data']
df = df[columns_to_keep]
```

```python
df
```

```python
import re

# Функция для извлечения значений из строки
def parse_instrument_data(data_str):
    if not isinstance(data_str, str):
        return {'min': None, 'avg': None, 'max': None, 'exp': None}
    
    min_val = re.search(r'min (-?\d+\.\d+)', data_str)
    avg_val = re.search(r'avg (-?\d+\.\d+)', data_str)
    max_val = re.search(r'max (-?\d+\.\d+)', data_str)
    exp_val = re.search(r'exp (-?\d+\.\d+)', data_str)
    
    return {
        'min': float(min_val.group(1)) if min_val else None,
        'avg': float(avg_val.group(1)) if avg_val else None,
        'max': float(max_val.group(1)) if max_val else None,
        'exp': float(exp_val.group(1)) if exp_val else None
    }

# Применение функции к колонке instrument_data и создание новых колонок
df[['min', 'avg', 'max', 'exp']] = df['instrument_data'].apply(parse_instrument_data).apply(pd.Series)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import re

# Собираем уникальные значения из колонки 'instrument'
unique_instruments = df['instrument'].unique()

# Построение и отображение корреляционных матриц для каждого уникального значения 'instrument'
for instrument in unique_instruments:
    subset = df[df['instrument'] == instrument]
    
    # Рассчитываем корреляционную матрицу
    correlation_matrix = subset[['profit', 'min', 'avg', 'max', 'exp']].corr()
    
    # Построение тепловой карты корреляционной матрицы
    plt.figure(figsize=(8, 6))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
    plt.title(f'Корреляционная матрица для инструмента: {instrument}')
    plt.show()
```
